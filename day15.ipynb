```python
import numpy as np
import matplotlib.pyplot as plt
import copy
```

```python
with open("/home/peter/Downloads/AdventOfCode/day15Input.txt") as fh:
    lines = fh.readlines()
#with open("/home/peter/Downloads/AdventOfCode/day15Test.txt") as fh:
#    lines = fh.readlines()
```

```python
def npFromLines(lines):
    board = np.zeros(shape=(len(lines),len(lines[0].strip())),dtype=np.int32)
    for i in range(len(lines)):
        for j in range(len(lines[i].strip())):
            board[i][j] = int(lines[i][j])
    return board
board = npFromLines(lines)

realBoard = np.zeros(shape=(board.shape[0]*5,board.shape[0]*5),dtype=np.int32)
for i in range(realBoard.shape[0]):
    for j in range(realBoard.shape[1]):
        baseVal = board[i%board.shape[0],j%board.shape[1]]
        distI = i//board.shape[0]
        distJ = j//board.shape[1]
        baseVal += distI + distJ
        baseVal = baseVal%9
        if baseVal == 0:
            baseVal = 9
        realBoard[i,j] = baseVal
board = realBoard
```

```python
print(np.max(realBoard))
```

```python
paths = np.zeros(shape=(board.shape[0],board.shape[1]),dtype=np.int32)
for el in range(len(paths)):
    paths[el] = 1000000000
paths[0,0] = 0



def getMinNeighbor(board,paths,point):
    x,y = point
    val = board[x,y]
    minSoFar = 10000000
    #print(point)
    if x > 0:
        tmp = val + paths[x-1,y]
        minSoFar = min(tmp,minSoFar)
     #   print(minSoFar)
        
    if y > 0:
        tmp = val + paths[x,y-1]
        minSoFar = min(tmp,minSoFar)
      #  print(minSoFar)
    if x+1 < board.shape[0]:
        tmp = val + paths[x+1,y]
        minSoFar = min(tmp,minSoFar)
       # print(minSoFar)
    if y+1 < board.shape[1]-1:
        tmp = val + paths[x,y+1]
        minSoFar = min(tmp,minSoFar)
        #print(minSoFar)
    #print(type(minSoFar))
    #print(type(paths[x,y]))
    if minSoFar < paths[x,y]:
        #print("changed ",paths[x,y]," to ", minSoFar)
        paths[x,y] = minSoFar
        return 1
    else: 
        #print("not changed on point",point)
        return 0

# return list of points within board neighboring the input
def getNeighbors(board,point):
    x,y = point
    points = []
    if x > 0:
        points.append((x-1,y))
    if y > 0:
        points.append((x,y-1))
    if x < board.shape[0]-1:
        points.append((x+1,y))
    if y < board.shape[1]-1:
        points.append((x,y+1))
    return points
    
    
updatedPoints = set()
updatedPoints.add((0,0))
while len(updatedPoints) > 0:
#if 1:
    newSet = set()
    for point in updatedPoints:
        newSet.update(getNeighbors(board,point))
    updatedPoints = set()
    for point in newSet:
        if getMinNeighbor(board,paths,point) == 1:
            updatedPoints.add(point)
    print("size of updated Points is ", len(updatedPoints))
    #viz = np.zeros(shape=(board.shape[0],board.shape[1]))
    #for point in updatedPoints:
    #    viz[point] = 1
    #    plt.imshow(viz)
```
